```python
import numpy as np
import matplotlib.pyplot as plt

# Define the training data
X = np.array([[-2, 4],
              [4, 1],
              [1, 6],
              [2, 4],
              [6, 2]])
y = np.array([-1, -1, 1, 1, 1])

# Initialize weights and bias
w = np.zeros(X.shape[1])
b = 0

# Define the perceptron function
def perceptron(X, y, w, b, learning_rate=0.1, num_epochs=100):
    for epoch in range(num_epochs):
        for i in range(X.shape[0]):
            z = np.dot(X[i], w) + b
            y_pred = 1 if z > 0 else -1
            if y_pred != y[i]:  # Update rule
                w += learning_rate * y[i] * X[i]
                b += learning_rate * y[i]
    return w, b

# Train the perceptron
w, b = perceptron(X, y, w, b)

# Plot the training data and decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
Z = np.sign(np.dot(np.c_[xx.ravel(), yy.ravel()], w) + b)
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.4, cmap=plt.cm.coolwarm)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, edgecolor='k')
plt.title('Perceptron Decision Region')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
